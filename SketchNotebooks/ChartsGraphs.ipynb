```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def minSec(sec):
    m = sec // 60
    s = sec % 60
    return f"{m:02d}:{s:02d}"
```

```python
df = pd.read_csv("./Shadowcasting.csv")
df
```

```python
%matplotlib inline
cm = 1/2.54  # centimeters in inches
```

```python
fig, ax = plt.subplots(figsize=(15*cm, 14*cm))
fig.set_dpi(600)


ax.plot(df['NumLights'], df['TT_ShadeShadow'], 'o-', color='b', ms=3, lw=0.75, fillstyle='none', label="Terrain Tools Shade+Shadow")
ax.plot(df['NumLights'], df['TT_ShadeOnly'], 'o--', color='b',  ms=2, lw=0.75, fillstyle='full', label="Terrain Tools Shade Only")


ax.plot(df['NumLights'], df['SN_ShadeShadow'], 'o-', color='g', ms=3, lw=0.75, fillstyle='none', label="Surface Normal Shade+Shadow")
ax.plot(df['NumLights'], df['SN_ShadeOnly'], 'o--', color='g',  ms=2,lw=0.75, label="Surface Normal Shade Only")
ax.plot(df['NumLights'], df['SN_ShadowOnly'], 'o:', color='g',  ms=2,lw=0.75,label="Surface Normal Shadow Only")

ytickmarks = [0, 300, 600, 900, 1200, 1500, 1800, 2100, 2400, 2700, 3000, 3300]
ax.set_yticks(ytickmarks)
ax.set_yticklabels([minSec(t) for t in ytickmarks], fontsize=6)
ax.set_ylabel('Time (mm:ss)')
ax.tick_params(axis='x', labelsize=6)
ax.set_xlabel('Number of Light Sources')
#ax.set_title('Sky Model Execution Times')
ax.legend(loc='upper left', fontsize=8, handlelength=3.75)
ax.grid('true')
```

```python
#
fig, ax = plt.subplots(figsize=(15*cm, 14*cm))
fig.set_dpi(600)

ax.plot(df['NumLights'], df['TT_ShadeShadow'] / df['NumLights'], 'o-', color='b',ms=3, lw=0.75, fillstyle='none', label="Terrain Tools Shade+Shadow")
ax.plot(df['NumLights'], df['TT_ShadeOnly'] / df['NumLights'], 'o--', color='b',ms=2, lw=0.75, label="Terrain Tools Shade Only")

ax.plot(df['NumLights'], df['SN_ShadeShadow'] / df['NumLights'], 'o-', color='g',ms=3, lw=0.75,fillstyle='none', label="Surface Normal Shade+Shadow")
ax.plot(df['NumLights'], df['SN_ShadeOnly']/ df['NumLights'], 'o--', color='g',ms=2, lw=0.75, label="Surface Normal Shade Only")
ax.plot(df['NumLights'], df['SN_ShadowOnly']/ df['NumLights'], 'o:', color='g',ms=2, lw=0.75, label="Surface Normal Shadow Only")
ax.set_yticks([i for i in range(0,14)])
ax.tick_params(axis='x', labelsize=6)
ax.tick_params(axis='y', labelsize=6)

ax.set_ylabel('Average Time per Hillshade (sec)')
ax.set_xlabel('Number of Light Sources')
#ax.set_title('Sky Model Execution Rate')
ax.grid(True)
ax.legend(loc='center right', fontsize=8, markerscale=0.9, handlelength=3.75)
plt.show()
```
